# Base plate (elevation end effector) workspace

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Simple open tree structure

```python
# Define the parameters of the kinematic chain
l1 = 350 # mm
l2 = 350 # mm
l3 = 50 # mm

# Joint angle ranges (in radians)
# Here we consider a 360 degree rotation for each joint
q1_range = np.linspace(0, np.pi/2, 30)
q2_range = np.linspace(-np.pi/2, np.pi/2, 30)
q3_range = np.linspace(-np.pi/2, np.pi/2, 30)
```

```python
# Compute and plot the end effector positions for all combinations of joint angles
x_vals = []
y_vals = []

for q1 in q1_range:
    for q2 in q2_range:
        for q3 in q3_range:
            theta1 = q1
            theta2 = q1 + q2
            theta3 = q1 + q2 + q3
            x = l1*np.cos(theta1) + l2*np.cos(theta2) + l3*np.cos(theta3)
            y = l1*np.sin(theta1) + l2*np.sin(theta2) + l3*np.sin(theta3)
            x_vals.append(x)
            y_vals.append(y)

plt.figure(figsize=(6,6))
plt.scatter(x_vals, y_vals, s=1, c='red')
plt.title("Workspace of Planar 2-Link Arm")
plt.xlabel("x")
plt.ylabel("z")
plt.axis('equal')
plt.grid(True)
plt.show()
```
